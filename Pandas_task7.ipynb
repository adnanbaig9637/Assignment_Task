### Deal with structural errors,

```python
df.head()
```

```python
df['InvoiceNo'] = df['InvoiceNo'].replace(r'C','',regex = True)
```

```python
df['InvoiceNo'] = df['InvoiceNo'].astype('int32')
```

```python
df['Quantity'] = df['Quantity'].replace(r'[-*# &]','',regex=True)
```

```python
df['Quantity'] = df['Quantity'].astype('int32')
```

```python
df['StockCode'] = df['StockCode'].replace(r'[ABCDEFGHIJHLMNOPQRSTUVWXYZ    ]','',regex=True)
```

### Standardization

```python
df.head()
```

```python
def standard(x):
    n = (x-df["UnitPrice"].mean())/(df["UnitPrice"].std())
    return n
df["UnitPrice"].apply(standard)
```

```python
def standard(x):
    n = (x-df["Quantity"].mean())/(df["Quantity"].std())
    return n
df["Quantity"].apply(standard)
```
